To merge the datasets from different sources, it is possible to combine them by multiple ways. A brief description is provided below. 

syntax: pd.merge(left, right, how, on)

Here, left and right indicate the first and second dataframe which are required to be merged

1. how = 'outer', This allows to take the union of datasets and append them together.
2. how = 'inner', This allows to take the intersection of datasets and keep only the common values
3. how = 'left', All indexes of left dataframe are included. If there are indexes which are not present in the 'left' dataframe, they will not be included in the merged dataset
4. how = 'right', All indexes of right dataframe are included. If there are indexes which are not present in the 'right' dataframe, they will not be included in the merged dataset

The keyword 'on' decides if the dataframes are merged on a particular ID which is a unique identifier. Eg. API Number in oil and gas wells. 

```python
# renaming column - Well name and Common well name
data2.rename(columns={"COMMON_WELL_NAME": "WELL NAME"}, inplace=True)
```

**Numpy - Scientific Computing**

This is a library in Python which is commonly used for scientific computing in Python, using multi-dimensional array objects. Creating a numpy array is a straight forward process. Examples of numpy arrays are shown in the section below along with multiple different methods which can be used for mathematical operations, slicing and indexing, sorting, manipulations etc. 

```python
# import the library
import numpy as np

array1 = np.array([1,2,3,4])
array2 = np.array([1,None,'Python',4.19])
array3 = np.array([[(1,3.14,6),(1,2,3)], [(100,200,300), (99,9.9,0.99)]])
array4 = np.arange(5, 20, 4) # evenly spaced values - step value
array5 = np.linspace(0,100,9) # evenly spaced values - number of samples
array6 = np.random.randint(10, size=6)  # One-dimensional array
```

```python
print(array1)
print(array2)
print(array3)
print(array4)
print(array5)
print(array6)
```

Numpy supports mathematical manipulations and a simple example is shown below. Regarding the complete list of possible functions, user can refer to the original documentation - https://numpy.org/devdocs/reference/routines.math.html. 


```python
# array mathematics - two different methods shown as example
print(array1 + array4)
print(np.add(array1, array4))
print(np.log(array1))

# sorting an array
array6.sort(axis=0)
print(array6)
```

```python
print("Number of Dimensions: ", array3.ndim)
print("Array shape:", array3.shape)
print("Array size: ", array3.size)
```

Accessing elements via indexing is very similar to a list but for a multi-dimensional array, elements can be accessed using a comma-separated tuple of indices. Slicing is also very similar. Some of the other more commonly used array manipulation routines are also available in numpy(). Entire list can be found out at https://numpy.org/devdocs/reference/routines.array-manipulation.html. Some examples are shown below.

```python
print(array2[2])
print(array3[0,0,1])
print(array4[1:4])
```

```python
i = np.append(array6, [25,26,27])
print(i)
concat = np.concatenate((array1, array2), axis=0)
print(concat)
mean_val = np.mean(array5)
print(mean_val)
```

```python
# Scatter Plot Example on Log-log scale

plt.figure(figsize=[8,8])

x = data['KLOGH (MD)']
y = data['SW (V/V)']

plt.scatter(x,y, s=5, label='SW') # s denotes the size of the bubbles in scatter plot

plt.xlabel('KLOGH (MD)', fontsize=15)
plt.ylabel('SW (V/V)', fontsize=15)
plt.title('Scatter Plot Example', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend(loc='best')
plt.xscale('log')
plt.yscale('log')
plt.xlim(10,10000)
plt.ylim(0.01, 10)
plt.show()
```
